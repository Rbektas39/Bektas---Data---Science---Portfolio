```python
# Bektas_IDS_Hwk6.ipynb
# Homework 6: California Housing Linear Regression
# ------------------------------------------------
# Run this top-to-bottom in a Jupyter notebook. Put interpretation answers in Markdown cells.

import numpy as np
import pandas as pd

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
import sys
print(sys.executable)
print(sys.version)
```

```python
# =========================
# Part 1: Data Loading + EDA
# =========================

housing = fetch_california_housing()

X = pd.DataFrame(housing.data, columns=housing.feature_names)
y = pd.Series(housing.target, name="med_house_value")

print("=== Part 1: Initial Exploration ===")

# (5 pts) Display first five rows (features + target together for convenience)
df = X.copy()
df["med_house_value"] = y
print("\nFirst 5 rows:")
display(df.head())
```

```python
# (5 pts) Feature names + missing values
print("\nFeature names:")
print(list(X.columns))
```

```python
print("\nMissing values per column:")
print(X.isna().sum())
```

```python
print("\nMissing values in target:")
print(y.isna().sum())
```

```python
# (10 pts) Summary statistics
print("\nSummary statistics (features):")
display(X.describe())

print("\nSummary statistics (target):")
display(y.describe())
```

```python
# ==================================
# Part 2: Linear Regression (Unscaled)
# ==================================


print("\n=== Part 2: Linear Regression on Unscaled Data ===")

# (5 pts) Train/test split (80/20)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

print(f"Train size: {X_train.shape}, Test size: {X_test.shape}")


```

```python
# (5 pts) Train model
lr_full = LinearRegression()
lr_full.fit(X_train, y_train)
```

```python
# (5 pts) Predict on test set
y_pred_full = lr_full.predict(X_test)
```

```python
# (5 pts) MSE
mse_full = mean_squared_error(y_test, y_pred_full)
```

```python
# (5 pts) RMSE
rmse_full = np.sqrt(mse_full)
```

```python
# (5 pts) R^2
r2_full = r2_score(y_test, y_pred_full)
```

```python
print("\nFull model performance (all features):")
print(f"MSE : {mse_full:.6f}")
print(f"RMSE: {rmse_full:.6f}")
print(f"R^2 : {r2_full:.6f}")
```

```python
# Coefficients for interpretation question
coef_full = pd.Series(lr_full.coef_, index=X.columns).sort_values(key=np.abs, ascending=False)
print("\nFull model coefficients (sorted by absolute value):")
display(coef_full.to_frame(name="coefficient"))

# Optional quick look: actual vs predicted (first 10)
compare_full = pd.DataFrame({"actual": y_test.values, "predicted": y_pred_full})
print("\nActual vs Predicted (first 10 rows of test set):")
display(compare_full.head(10))
```

```python
# ==================================================
# Part 3: Feature Selection + Simplified 3-feature model
# ==================================================


print("\n=== Part 3: Feature Selection and Simplified Model ===")

# Choose top 3 features by absolute coefficient magnitude (data-driven, reproducible)
selected_features = list(coef_full.index[:3])
print("Selected 3 features (largest absolute coefficients from full model):")
print(selected_features)

X3 = X[selected_features]

# Split again for fairness (same random_state)
X3_train, X3_test, y3_train, y3_test = train_test_split(
    X3, y, test_size=0.20, random_state=42
)

```

```python
# (5 pts) Train simplified model
lr_3 = LinearRegression()
lr_3.fit(X3_train, y3_train)

# Predict
y_pred_3 = lr_3.predict(X3_test)
```

```python
# (5 pts) Evaluate simplified model
mse_3 = mean_squared_error(y3_test, y_pred_3)
rmse_3 = np.sqrt(mse_3)
r2_3 = r2_score(y3_test, y_pred_3)
```

```python
print("\nSimplified model performance (3 features):")
print(f"MSE : {mse_3:.6f}")
print(f"RMSE: {rmse_3:.6f}")
print(f"R^2 : {r2_3:.6f}")

# Compare side-by-side
perf = pd.DataFrame(
    {
        "Model": ["Full (all features)", "Simplified (3 features)"],
        "MSE": [mse_full, mse_3],
        "RMSE": [rmse_full, rmse_3],
        "R2": [r2_full, r2_3],
    }
)
print("\nPerformance comparison:")
display(perf)

# Coefficients for simplified model
coef_3 = pd.Series(lr_3.coef_, index=selected_features).sort_values(key=np.abs, ascending=False)
print("\nSimplified model coefficients:")
display(coef_3.to_frame(name="coefficient"))
```
